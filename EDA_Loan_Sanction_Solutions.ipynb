```python
import pandas as pd
pd.set_option('display.max_columns',300)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data=pd.read_excel('loan_sanction.xlsx')
data
```

```python
data.info
```

```python
data.columns
```

```python
data = pd.DataFrame(data, columns= ['Loan_ID', 'Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area'])
data
```

```python
dfresult = pd.DataFrame(data)
dfresult
```

```python
print(dfresult)
```

```python
##Check how many duplicate rows are present
duplicaterows = (dfresult.duplicated==True)


duplicaterows
```

```python
##If any Duplicates Present, drop those duplicates
dfresult.drop_duplicates()
dfresult.count()
```

```python
##Check null values in each column
dfresult.isnull().sum()
```

```python
##Delete the Rows which are having null values.
dfresult = dfresult.dropna()
dfresult.count()
```

```python
##Impute null values with Mode in Gender Column
dfresult.fillna(dfresult['Gender'].mode()[0], inplace = True)
print(dfresult)
```

```python
##Impute null values with Mode in Self_Employed Column
dfresult.fillna(dfresult['Self_Employed'].mode()[0], inplace = True)
dfresult
```

```python
##Impute null values with Median in LoanAmount Column
dfresult.fillna(dfresult['LoanAmount'].median(), inplace = True)
dfresult
```
