```python
pd.set_option('display.max_columns', 300)
STI=pd.read_excel('EDA_Shark_Tank_India.xlsx')

STI
```

```python
#Find the number of episodes?
print(STI['episode_number'].nunique())
```

```python
#Max, min , mean of asked amount,asked equity, asked valuation? 
Asked_amt_min = STI["pitcher_ask_amount"].min()
Asked_amt_max = STI["pitcher_ask_amount"].max()
Asked_amt_mean = STI["pitcher_ask_amount"].mean()
print("Min of asked amount:", Asked_amt_min)
print("Max of asked amount:", Asked_amt_max)
print("Mean of asked amount:", Asked_amt_mean)

Asked_equity_min = STI["ask_equity"].min()
Asked_equity_max = STI["ask_equity"].max()
Asked_equity_mean = STI["ask_equity"].mean()

print("Min of asked equity:", Asked_equity_min)
print("Max of asked equity:", Asked_equity_max)
print("Mean of asked equity:", Asked_equity_mean)

Asked_valuation_min = STI["ask_valuation"].min()
Asked_valuation_max = STI["ask_valuation"].max()
Asked_valuation_mean = STI["ask_valuation"].mean()

print("Min of asked valuation:", Asked_valuation_min)
print("Max of asked valuation:", Asked_valuation_max)
print("Mean of asked valuation:", Asked_valuation_mean)
```

```python
#Max & Min asked- equity,asked-valuation and asked amount episode-wise?
max_min_values = STI.groupby('episode_number').agg({
    'ask_equity': ['max', 'min'],
    'ask_valuation': ['max', 'min'],
    'pitcher_ask_amount': ['max', 'min']
}).reset_index()

max_min_values
```

```python
#Episode wise minimum and maximum of sharks invested?
episode_wise = STI.groupby("episode_number").agg({
    "pitcher_ask_amount": ["max", "min"],
   
})
episode_wise
```

```python
#Brand names in which 2 ,3 or 4 sharks are invested? 
Brandnames1=STI[(STI['total_sharks_invested'] ==2) | (STI['total_sharks_invested'] ==3) | (STI['total_sharks_invested'] ==4 )][['brand_name','total_sharks_invested']]

Brandnames1 = Brandnames1.sort_values(by='total_sharks_invested', ascending=False)
Brandnames1
```

```python
# Episode wise minimum and maximum of sharks invested? 
invested_sharks = STI.groupby('episode_number')['total_sharks_invested'].agg(['min', 'max'])

invested_sharks
```

```python
#Brand names who ask for 1 crore and got a deal? 
Brandnames_with_deal = STI[(STI['pitcher_ask_amount'] == 1000) & (STI['deal'] == '1')][['brand_name']]

print("Brand names that asked for 1 crore and got a deal:")

Brandnames_with_deal
```

```python
#Brand names where deal equity is between 0 to 50? 
Brandname=STI[(STI['deal_equity'] >0) & (STI['deal_equity']<50)][['brand_name','deal_equity']].reset_index()
Brandname = Brandname.sort_values(by='deal_equity', ascending=False)
Brandname
```

```python
#Find the number of brands participated in each episode? 
brands_per_episode = STI.groupby('episode_number')['brand_name'].nunique().reset_index()
brands_per_episode.columns = ['episode_number', 'brands_participated']
print(brands_per_episode)
```

```python
#How many sharks participated in this show and What were their names? 
shark_names = [col.replace('_present', '').capitalize() for col in STI.columns if col.endswith('_present')]
print(shark_names)
```

```python
#Find appearance of each sharks? 

shark_appearances = {}
for column_name in STI.columns:
    if "_present" in column_name:
        shark_name = column_name.replace("_present", "").capitalize()
        appearances = STI[column_name].sum()  
        shark_appearances[shark_name] = appearances
        
shark_appearances_df = pd.DataFrame(list(shark_appearances.items()), columns=['Shark_Name', 'Appearance_Count'])
shark_appearances_df
        
```

```python
#How many entrepreneurs were present? 
Entrepreneurs=len(STI['brand_name'])
Entrepreneurs
```

```python
#Find the equity percent that each sharks gets?

shark_columns = ['ashneer_deal', 'anupam_deal', 'aman_deal', 'namita_deal', 'vineeta_deal','peyush_deal','ghazal_deal']

shark_equity_total = {}

for column in shark_columns:
    shark_name = column.replace('_deal', '').capitalize()
    total_equity_for_shark = STI[column].sum()
    shark_equity_total[shark_name] = total_equity_for_shark
total_equity_offered = STI['equity_per_shark'].sum()

shark_equity_percent = {}

for shark, total_equity in shark_equity_total.items():
    equity_percent = (total_equity / total_equity_offered) * 100
    shark_equity_percent[shark] = equity_percent
    
shark_equity_percent
```
